```python
df.columns = df.columns.str.strip() 

df.columns
```

```python
# Selecting the required variables including the target variable
selected_columns = [
    'Life_expectancy', 'Alcohol', 'Adult_Mortality', 'Hepatitis_B', 'Measles', 
    'BMI', 'Polio', 'Total_expenditure', 'Diphtheria', 'HIV/AIDS', 'GDP', 
    'thinness_1-19_years', 'Schooling', 'infant_deaths'
]

# Creating a new dataframe with the selected variables
selected_data = df[selected_columns]

# Checking for missing values in the selected variables
missing_data = selected_data.isnull().sum()
missing_data_summary = missing_data[missing_data > 0]

missing_data_summary
```

```python
# Filling missing values with the median for each column
for column in missing_data_summary.index:
    median_value = selected_data[column].median()
    selected_data[column].fillna(median_value, inplace=True)

# Confirming no more missing values
selected_data.isnull().sum().max()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Prepare the features and target
X = selected_data.drop('Life_expectancy', axis=1)
y = selected_data['Life_expectancy']

# Splitting the dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Creating the linear regression model
model = LinearRegression()

# Training the model
model.fit(X_train, y_train)

# Predicting on the test set
y_pred = model.predict(X_test)

# Calculating the performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, r2
```
